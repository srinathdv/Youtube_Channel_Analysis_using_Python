## Data Cleaning

```python
df.head(20)
```

```python
#replace -- with NaN
df = df.replace('--',np.nan,regex=True)
df.head(20)
```

```python
df.isnull().sum()
```

```python
df=df.fillna(0)
df.isnull().sum()
```

```python
df.dtypes
```

```python
#lets change the Rank to int
#first remove the string from the Rank column
df['Rank']=df['Rank'].str[0:-2]   #t @ -1 position, s @-2 position
df
```

```python
#Lets remove the comma from the Rank Column
df['Rank']= df['Rank'].str.replace(',','')
df
```

```python
#convert video uploads and subscribers column to int
df['Video Uploads']=df['Video Uploads'].astype(int)
df['Rank'] = df['Rank'].astype(int)
df['Subscribers']  = df['Subscribers'].astype(int)
df.dtypes
```

```python
#lets clean Grade column
df['Grade'].unique()
```

```python
#lets map Grade to Numeric Value
df['Grade'] = df['Grade'].map({'A++ ':5,'A+ ':4,'A ':3,'A- ':2,'B+ ':1})
df
```

```python
df.dtypes
```

## Average Views For Each Channel

```python
df.columns
```

```python
#create new Column having Avg views
df['Avg_Views'] = df['Video views']/df['Video Uploads']
df
```

## Top Five Channels With Maximum Number of Video Uploads

```python
df.columns
```

```python
df.sort_values(by='Video Uploads',ascending=False).head()
```

## Correlation Matrix

```python
df.dtypes
```

```python
#select only numeric value Columns
numeric_columns = df.select_dtypes(include=[np.number])
numeric_columns
```

```python
#calculate correlation now
correlation_matrix = numeric_columns.corr()
correlation_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix,annot=True,fmt = '.2f')
plt.title('Correlation Matrix')
plt.show()
```

## Which Grade Has A Maximum Number of Video Uploads?

```python
max_videos = df.groupby('Grade')['Video Uploads'].mean().sort_values(ascending=False)
max_videos
```

```python
sns.barplot(x='Grade',y='Video Uploads',data=df)
plt.title('Grade VS Video Uploads')
plt.show()
```

* From the above graph it is clear that when the grade is high, video upload is also high. Video Uploads are high in 'A++' video channels.

## Which Grade Has The Highest Average Views?

```python
Highest_avg_views = df.groupby('Grade')['Avg_Views'].mean().sort_values(ascending=False)
Highest_avg_views
```

```python
sns.barplot(x='Grade',y='Avg_Views',data=df)
plt.title('Grade VS Average views')
plt.show()
```

* Channels with 'A+' grade has higher number of average views compared to others.

## Which Grade Has The Highest Number of Subscribers? 

```python
Highest_Subs = df.groupby('Grade')['Subscribers'].mean().sort_values(ascending=False)
Highest_Subs
```

```python
sns.barplot(x='Grade',y='Subscribers',data=df)
plt.title('Grade VS Subscribers')
plt.show()
```

From the above graph it is clear that channel with the 'A++' Grade has the highest number of subscribers
